```python
import json
```

```python
agi_eval = 'AGI_EVAL.json'
emma_math = 'emma_math_dataset.json'
emma_physics = 'emma_physics_dataset.json'
gpqa = 'gpqa_physics_dataset.json'
gsm8k = 'gsm8k_dataset.json'
jeebench = 'jeebench_dataset.json'
mathvista = 'mathvista_dataset.json'
mmlu_math = 'mmlu_math_physics_test.json'
mmmu_arch = 'mmmu_arch_eng_test_dataset.json'
olympiad = 'olympiad_bench_dataset.json'
scibench_data = 'scibench_dataset.json'
scienceqa = 'scienceqa_test_dataset.json'
scieval_physics = 'scieval_physics_combined.json'
```

```python
with open(agi_eval, 'r') as f:
    agi_eval_data = json.load(f)
with open(emma_math, 'r') as f:
    emma_math_data = json.load(f)
with open(emma_physics, 'r') as f:
    emma_physics_data = json.load(f)
with open(gpqa, 'r') as f:
    gpqa_data = json.load(f)
with open(gsm8k, 'r') as f:
    gsm8k_data = json.load(f)
with open(jeebench, 'r') as f:
    jeebench_data = json.load(f)    
with open(mathvista, 'r') as f:
    mathvista_data = json.load(f)
with open(mmlu_math, 'r') as f:
    mmlu_math_data = json.load(f)
with open(mmmu_arch, 'r') as f:
    mmmu_arch_data = json.load(f)
with open(olympiad, 'r') as f:
    olympiad_data = json.load(f)
with open(scibench_data, 'r') as f:
    scibench_data = json.load(f)
with open(scienceqa, 'r') as f:
    scienceqa_data = json.load(f)
with open(scieval_physics, 'r') as f:
    scieval_physics_data = json.load(f)

datasets = {
    'AGI_EVAL': agi_eval_data,
    'EMMA_MATH': emma_math_data,
    'EMMA_PHYSICS': emma_physics_data,
    'GPQA': gpqa_data,
    'GSM8K': gsm8k_data,
    'JEEBENCH': jeebench_data,
    'MATHVISTA': mathvista_data,
    'MMLU_MATH': mmlu_math_data,
    'MMMU_ARCH': mmmu_arch_data,
    'OLYMPIAD': olympiad_data,
    'SCIBENCH': scibench_data,
    'SCIENCEQA': scienceqa_data,
    'SCIEVAL_PHYSICS': scieval_physics_data
}
```

```python
total_questions = 0
total_image_based = 0
for i in datasets:
    print(f"Dataset: {i}")
    images_based = 0
    for j in datasets[i]:
        if 'image' in j:
            if len(j['image']) > 0:
                images_based += 1
    print(f"Number of images based questions: {images_based}")
    print(f"Total number of questions: {len(datasets[i])}")
    print(f"Percentage of images based questions: {images_based/len(datasets[i])*100:.2f}%")
    print(f"Percentage of text based questions: {(len(datasets[i]) - images_based)/len(datasets[i])*100:.2f}%")
    total_questions += len(datasets[i])
    total_image_based += images_based
    print()
print(f"Total number of questions: {total_questions}")
print(f"Total number of images based questions: {total_image_based}")
print(f"Total number of text based questions: {total_questions - total_image_based}")
print(f"Percentage of images based questions: {total_image_based/total_questions*100:.2f}%")
print(f"Percentage of text based questions: {(total_questions - total_image_based)/total_questions*100:.2f}%")
```
